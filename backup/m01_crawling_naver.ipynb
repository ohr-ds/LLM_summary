# 검색 설정

```python
# 크롤링 대상
target = "blog", "cafearticle"

# 파일 형태
format = "json"

# 검색 결과 개수(default: 10, max: 100)
display = 100

# 검색 결과 정렬(sim: 정확도순으로 내림차순 정렬(기본값) date: 날짜순으로 내림차순 정렬)
sort = "sim"

# 검색어 정보
query = ["스탠바이미", "스탠바이미고"]
```

# 크롤링 url 생성

```python
"""
target = "blog", "cafearticle"
format = "json", "xml"
"""
# 검색어 인코딩
encText = urllib.parse.quote(query[0])

# url 생성
url = api_address + target[0] + "." + format[0] + "?query=" + encText + "&display=" + str(display) + "&sort=" + sort
print(url)
```

# 호출& 결과

```python
# 호출
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

# 결과
response = urllib.request.urlopen(request)
rescode = response.getcode()
try:
    response_body = response.read()
    result = response_body.decode('utf-8')
    print("Success")
except: # todo: set logging
    logging.error("Error Code:" + rescode)
    # print("Error Code:" + rescode)
```
