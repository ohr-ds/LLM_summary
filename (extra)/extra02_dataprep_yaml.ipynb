# 제외어

```python
def contains_exclude_terms(description, phrases, hashtags):
    for phrase in phrases:
        if phrase in description:
            return True
    for hashtag in hashtags:
        if hashtag in description:
            return True
    return False
```

```python
# 제외 확인
df_null.loc[df_null['description'].apply(lambda desc: contains_exclude_terms(desc, exclude_phrase, exclude_hashtag))]
```

```python
# Filtering the DataFrame
df_exclude = df_null[~df_null['description'].apply(lambda desc: contains_exclude_terms(desc, exclude_phrase, exclude_hashtag))]
```

```python
df_exclude.reset_index(drop=True, inplace=True)
```
